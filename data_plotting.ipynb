```python
import matplotlib.pyplot as plt
from matplotlib import style

style.use('ggplot')

#prepare the data 
x = [1, 2, 3, 4, 5]
y1 = [0, 20, 30, 25, 15]
y2 = [0, 5, 12, 27, 35]

#plot the data 
plt.scatter(x, y1, label='Mild Cheddar', color='teal')
plt.scatter(x, y2, label='Blue Thunder', color='lightcoral')

#add a legend 
plt.legend()

#add a title and axis labels
plt.title("cheese flavour over time")

plt.xlabel("cheese age(years)")
plt.ylabel("tastiness score")

# show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np

style.use('ggplot')

#prepare the data. use numpy so we can perform arithmetic on the whole list once
#x = [1, 2, 3, 4, 5]
x = np.array([1, 2, 3 ,4 ,5])
y1 = [0, 20, 30, 25, 15]
y2 = [0, 5, 12, 27, 35]

bar_width = 0.4



#plot the data 
plt.bar(x-bar_width/2, y1, label='Mild Cheddar', width= bar_width, color='teal')
plt.bar(x+bar_width/2, y2, label='Blue Thunder',width= bar_width, color='lightcoral')

#add a legend 
plt.legend()

#add a title and axis labels
plt.title("cheese flavour over time")

plt.xlabel("cheese age(years)")
plt.ylabel("tastiness score")

# show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# prepare the data 
x = [1, 2, 3, 4, 5]
y1 = [0, 20, 30, 25, 15]
y2 = [0, 5, 12, 27, 35]
y3 = [0, 2, 7, 12, 35]
y4 = [0, 4, 6, 12, 12]
y5 = [0, 5, 6, 8, 10]
y6 = [0, 6, 12, 15, 19]

# create the figure that will contain the axes
fig = plt.figure(figsize=(12, 8))

plt.subplot(231)
plt.plot(x, y1, label='Mild Cheddar')
plt.legend()

plt.subplot(232)
plt.plot(x, y2, label='Blue Thunder')
plt.legend()

plt.subplot(233)
plt.plot(x, y3, label='Gorgonzola')
plt.legend()

plt.subplot(234)
plt.plot(x, y4, label='Brie')
plt.legend()

plt.subplot(235)
plt.plot(x, y5, label='Manchego')

# show the plot
plt.show()
```

```python
# to confirm that the figure is made up of a number of axes
fig.get_axes()
```

```python
# method no 2
# prepare the data 
x = [1, 2, 3, 4, 5]
y1 = [0, 20, 30, 25, 15]
y2 = [0, 5, 12, 27, 35]
y3 = [0, 2, 7, 12, 35]
y4 = [0, 4, 6, 12, 12]
y5 = [0, 5, 6, 8, 10]
y6 = [0, 6, 12, 15, 19]

# create a grid with 2 rows and 3 columns
fig, ax = plt.subplots(2,3, figsize=(12,8))


ax[0,0].plot(x, y1, label='Mild Cheddar')
ax[0,0].legend()

ax[0,1].plot(x, y2, label='Blue Thunder')
ax[0,1].legend()


ax[0,2].plot(x, y3, label='Gorgonzola')
ax[0,2].legend()

ax[1,0].plot(x, y4, label='Brie')
ax[1,0].legend()


ax[1,1].plot(x, y5, label='Manchego')
ax[1,1].legend()

# show the plot
plt.show()
```

## spanning rows and columns

sometimes we want figures to span across a number of rows or columns in our grid. we can use subplot2grid() for this

```python
# number of rows and columns
gridsize = (3, 2)

# create the figure that will contain the axes
fig = plt.figure(figsize=(12, 8))

# create an axes that spans 2 rows and 2 columns
ax1 = plt.subplot2grid(gridsize, (0,0), colspan=2, rowspan=2)

# create two axes that occupy the other 2 positions in the grid
ax2 = plt.subplot2grid(gridsize, (2,0))
ax3 = plt.subplot2grid(gridsize, (2,1))
```
